```python
import pandas as pd

# Employee details
emp = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Saurav', 'Rahul', 'Riya', 'Sneha'],
    'dept_id': [101, 102, 103, 101]
})

# Department details
dept = pd.DataFrame({
    'dept_id': [101, 102, 104],
    'dept_name': ['IT', 'Finance', 'HR']
})

```

```python
pd.merge(emp,dept,on = "dept_id",how='inner')
```

```python
pd.merge(emp,dept,on = "dept_id",how='left')
```

```python
pd.merge(emp,dept,on = "dept_id",how='right')
```

```python
pd.merge(emp,dept,on = "dept_id",how='outer')
```

```python
data = pd.DataFrame({
    'department': ['IT', 'Finance', 'IT', 'HR', 'Finance', 'HR'],
    'city': ['Kolkata', 'Delhi', 'Pune', 'Mumbai', 'Kolkata', 'Delhi'],
    'salary': [50000, 60000, 55000, 62000, 58000, 59000],
    'age': [25, 32, 28, 29, 26, 31]
})
```

```python
pd.pivot_table(data, values='salary', index='department', aggfunc='mean')
```

```python
pd.pivot_table(data, values=['salary','age'], index='department', aggfunc={'salary':'mean','age':'max'})
```
